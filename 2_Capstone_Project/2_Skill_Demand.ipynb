## WHAT IS THE MOST DEMAND SKILL FOR THE TOP 3 POPULAR DATA ROLES

Methodology

1. Clean up skill column
2. Calculate Skill based on job_title_short
3. Calculate skill percetage
4. Plot final findings

```python
# Importing libraries
import ast
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Loading Datasets

df = pd.read_csv(r"C:\Users\User\Documents\project_data\SampleData\data_jobs.csv")
```

```python
# Data Cleaning
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_us = df[df['job_country'] == 'United States']

df_us
```

```python
df_skills = df_us.explode('job_skills')
df_skills[['job_title', 'job_skills']]
```

```python
df_skills_count = df_skills.groupby(['job_skills', 'job_title_short']).size()

df_skills_count = df_skills_count.reset_index(name='skill_count')

df_skills_count.sort_values(by='skill_count', ascending=False, inplace=True)

df_skills_count
```

```python
job_titles = df_skills_count['job_title_short'].unique().tolist()

job_titles = sorted(job_titles[:3])

job_titles
```

```python
fig, ax = plt.subplots(len(job_titles), 1)

for i,  job_title in enumerate(job_titles):
    df_plot = df_skills_count[df_skills_count['job_title_short'] == job_title].head(5)
    df_plot.plot(kind='barh', x='job_skills', y='skill_count', ax=ax[i], title=job_title)
    ax[i].invert_yaxis()
    ax[i].set_ylabel('')
    ax[i].legend().set_visible(False)

fig.suptitle('Count on Top Skills in Job Posting', fontsize=15)
fig.tight_layout(h_pad=0.5)
plt.show()
```

```python
df_job_title_count = df_us['job_title_short'].value_counts().reset_index(name='jobs_total')

df_job_title_count
```

```python
df_skills_perc = pd.merge(df_skills_count, df_job_title_count, how='left', on='job_title_short')

df_skills_perc['skill_percentage'] = 100 * df_skills_perc['skill_count'] / df_skills_perc['jobs_total']

df_skills_perc
```

```python
fig, ax = plt.subplots(len(job_titles), 1)

sns.set_theme(style='ticks')

for i,  job_title in enumerate(job_titles):
    df_plot = df_skills_perc[df_skills_perc['job_title_short'] == job_title].head(5)
    sns.barplot(data=df_plot, x='skill_percentage', y='job_skills', ax=ax[i], palette='dark:b_r')
    ax[i].set_title(job_title)
    ax[i].set_ylabel('')
    ax[i].set_xlabel('')
    ax[i].get_legend()
    ax[i].set_xlim(0, 78)

    for n, v in enumerate(df_plot['skill_percentage']):
        ax[i].text(v + 1, n, f'{v:.0f}%', va='center')
    
    if i != len(job_titles) - 1 :
        ax[i].set_xticks([])

fig.suptitle('Likelihood Request of Skills in Job Posting', fontsize=15)
fig.tight_layout(h_pad=0.5)
plt.show()
```
