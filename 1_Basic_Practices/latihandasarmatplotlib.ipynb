```python
# import library
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load dataset nya
df = pd.read_csv(r"C:\Users\User\Documents\project_data\SampleData\data_jobs.csv")
```

```python
# Data cleaning bagian tanggal, lebih ke ngubah format aja sih
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# Grouping
df.groupby('job_title_short')['salary_year_avg'].median().sort_values()
```

```python
job_salary = df.groupby('job_title_short')['salary_year_avg'].median().sort_values()
```

```python
job_salary.plot(kind='barh')
```

```python
# modify tampilan agar lebih mantulululul
job_salary.plot(kind='barh')
plt.xlabel('Gaji dalam $ USD')
plt.ylabel('')
plt.title('Gaji Rata-Rata Tahunan Bidang Data')
plt.show()
```
