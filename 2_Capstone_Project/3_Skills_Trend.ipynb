## How are in-demand skills trending for data analysts?

**Methodology**
1. Aggregate skill count monthly
2. Re-analyze based on percentage of total jobs
3. Plot the monthly skill on demand

```python
# Importing libraries
import ast
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Loading Datasets

df = pd.read_csv(r"C:\Users\User\Documents\project_data\SampleData\data_jobs.csv")
```

```python
# Data Cleaning
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA_US = df[(df['job_title'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()
```

```python
df_DA_US['job_posted_month_no'] = df_DA_US['job_posted_date'].dt.month
```

```python
df_DA_US_explode = df_DA_US.explode('job_skills')
```

```python
df_DA_US_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)
```

```python
df_DA_US_pivot = df_DA_US_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)

df_DA_US_pivot.loc['Total'] = df_DA_US_pivot.sum()

df_DA_US_pivot[df_DA_US_pivot.loc['Total'].sort_values(ascending=False).index]
```

```python
df_DA_US_pivot = df_DA_US_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)

df_DA_US_pivot.loc['Total'] = df_DA_US_pivot.sum()

df_DA_US_pivot = df_DA_US_pivot[df_DA_US_pivot.loc['Total'].sort_values(ascending=False).index]

df_DA_US_pivot = df_DA_US_pivot.drop('Total')

df_DA_US_pivot
```

```python
df_DA_US.groupby('job_posted_month_no').size()
```

```python
Df_DA_Totals = df_DA_US.groupby('job_posted_month_no').size()
```

```python
df_DA_US_Percent = df_DA_US_pivot.div(Df_DA_Totals/100, axis=0)

df_DA_US_Percent
```
